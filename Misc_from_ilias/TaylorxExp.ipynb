```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial as atp
#
def f(x):
    return x*np.exp(x)
#

x = np.linspace(0,3,100)
fig, ax = plt.subplots()
ax.set_title(r'Taylor expansion of $f(x)=x\cdot e^x$ to different orders')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
ax.plot(x, x*np.exp(x), label = r'$x\cdot e^x$')

taylor = atp(f,0,3,3) # 3th Taylor pol
ax.plot(x, taylor(x), label="order {}".format(3))

ax.legend()
```
