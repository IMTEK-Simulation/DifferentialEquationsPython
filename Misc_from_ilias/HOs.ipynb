# The Harmonic Oscillator

Hier und im folgen handelt es sich immer um das Gleichungssystem
\begin{align}
\dot{y}_1&=y_2\\
\dot{y}_2&=-y_1-f(t,y_1,y_2)
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def dydt(t,r):
    y1, y2 = r
    dy1dt = y2
    dy2dt = f(t,y1,y2) # Let us insert here a general function
    return dy1dt, dy2dt
```

## H.O. with and without damping

```python
# Change this definition according to the requirements of the problem set
def f(t,y1,y2):
    gamma = 0.2  # Hier an der DÃ¤mpfung drehen
    amp = 0.25    # Hier an der Amplitude drehen
    omega = 0.979
    return -y1 - gamma*y2 - amp*np.cos(omega*t)  
#
t0 = 0.
t1 = 100.
n = 10000
t_span = [t0, t1]
t_eval = np.linspace(t0,t1,n)
y0 = [1.0, 0.]
sol = solve_ivp(dydt, t_span, y0, t_eval= t_eval)
y1 = sol.y[0]
y2 = sol.y[1]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.tight_layout()
#
ax1.plot(y1,y2)
ax1.set_xlabel(r'$y_1$')
ax1.set_ylabel(r'$y_2$')
#
y1span = np.abs(y1.max()-y1.min())
y2span = np.abs(y2.max()-y2.min())
ax1.set_aspect(y1span/y2span)
#
ax2.plot(t_eval,sol.y[0], label=r'$y_1$')
ax2.plot(t_eval,sol.y[1], label=r'$y_2$')
ax2.set_xlabel(r'$t$')
ax2.set_ylabel(r'$y_2,\,y_2$')
ax2.set_aspect(100./np.maximum(y1span,y2span))
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.legend()
```

### Tunneling diode

Hier Ã¤ndert sich auch das Gleichungssystem. Wir haben
\begin{align}
\frac{dU}{dt}&=\frac{1}{C}(I - I_T(U)) \\
\frac{dI}{dt}&=\frac{1}{L}(U_{in} - R \cdot I -U)
\end{align}

```python
# Change this definition according to the requirements of the problem set
def IT(U):
    return 2*U**3-U**2+0.13*U
#
def dydt(t,r):
    Uin = .22
    R = 10.
    C = 1.e-4
    L = 0.1
    U, I = r
    dUdt = (I-IT(U))/C
    dIdt = (Uin-R*I-U)/L    # No more general function
    return dUdt, dIdt
```

```python
#
t0 = 0.
t1 = 1.
n = 1000
t_span = [t0, t1]
t_eval = np.linspace(t0,t1,n)
y0 = [0.0, 0.0]
sol = solve_ivp(dydt, t_span, y0, t_eval= t_eval)
U = sol.y[0]
I = sol.y[1]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.tight_layout()
#
ax1.plot(U,I)
ax1.set_xlabel('U')
ax1.set_ylabel('I')
#
ax2.plot(t_eval[900:],U[900:])
ax2.set_xlabel('Time t')
ax2.set_ylabel('U')
ax2.set_aspect(1./3.)
#
ax3.plot(t_eval[900:],I[900:])
ax3.set_xlabel('Time t')
ax3.set_ylabel('I')
ax3.set_aspect(0.1/0.008)
```
